```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
df = pd.read_csv('sro_data_examples/danila_dance.txt')
df_filtered = df.query('Distance<1000')
df_filtered      
```

```python
import matplotlib.pyplot as plt

# gca stands for 'get current axis'
ax = plt.gca()

df_filtered.plot(kind='line',x='Time',y='Distance',ax=ax)
plt.show()
```
